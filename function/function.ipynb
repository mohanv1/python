```python
# how to acess local varible outside the function

def add_sub(a,b):
    c=a+b
    d=a-b
    return c,d
added_value,subtracted_value=add_sub(20,20)    
print("added_value",added_value)
print("subtracted_value",subtracted_value)
```

```python
#how to change the global variable to local inside the function using global keyword

name="Mohan"
def na():
    global name
    name="python"
    print("Inside the function:",name)
na()
print("Outside the function",name)
    
```

```python
#how to print the global variable inside the function

name="Mohan"
def na():
    name="python"
    print("Inside the function:",name)
    print("Inside the function with globals:",globals()['name'])
na()
print("Outside the function",name)
    
```

```python
# how to check odd or even using function 

def check(l):
    for i in l:
        if i%2==0:
            print("The number is even:",i)
        else:
            print("The number is odd",i)
l=[2,7,9,6,55,2]
check(l)

print('----------------')

def check(l):
    odd=0
    even=0
    for i in l:
        if i%2==0:
            even+=1
        else:
            odd+=1
    return even,odd        
l=[2,7,9,6,55,29]
even,odd=check(l)
print("Even count is:",even)
print("Odd count is:",odd)
```

```python
# how to change the global value inside the function 

name="Mohan"
def na():
    name="python"
    print("Inside the function:",name)
    print("Inside the function with globals:",globals()['name'])
    globals()['name']='coder'
na()
print("Outside the function",name)
    
```

```python
#arguments
#positional arguements

def bio(name,age):
    print("Name is:",name)
    print("Age is:",age)
bio("mohan","18") 

#keyword arguments
def bio(age,name):
    print("Name is:",name)
    print("Age is:",age)
bio(name="mohan",age="18")    
```

```python
#default arguements

def bio(name,age='18'):
    print("Name is:",name)
    print("Age is:",age)
bio(name="mohan") 
```

```python
#variable length arguements

def sum(x,*y):
    s=x
    for i in y:
        s=s+i
    print("The added value is:",s)
sum(10,20,30,40)    
        
    
```

```python
#kwargs **

#1
def bio(name,**additional):
    print("Name:",name)
    for i,j in additional.items():
        print(i,':',j)
bio(name="mohan",age='17',city="tiruppur",college='jsrec')

print('----------------')

#2
def bio(name,**additional):
    print("Name:",name)
    print("age:",additional['age'])
bio(name="mohan",age='17',city="tiruppur",college='jsrec')
```

```python
#recursive fuction
def rec(n):
    if n==0:
        return 1
    return n*rec(n-1)
fac=rec(5)
print("Factorial of 5 is :",fac)
```
